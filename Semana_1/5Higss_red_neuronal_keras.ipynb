# 1- El problema a resolver

Los fГӯsicos usan aceleradores de partГӯculas, como el [Gran Colisionador de Hadrones](https://es.wikipedia.org/wiki/Gran_colisionador_de_hadrones) ubicado en la frontera entre Suiza y Francia, para descubrir nuevas partГӯculas subatГіmicas.

La detecciГіn de estas partГӯculas usualmente es un problema de clasificaciГіn, para determinar si el dato observado corresponde a la seГұal de interГ©s o al ruido de fondo.

La idea es desarrollar un modelo de Deep Learning capaz de realizar esta clasificaciГіn.

# 2- El set de datos

El set de datos fue desarrollado por investigadores del Departamento de FГӯsica y AstronomГӯa de la Universidad de California Irvine.

En total contiene 11 millones de datos, cada uno representado por 28 caracterГӯsticas (*features*):

- Las primeras 21 corresponden a propiedades cinemГЎticas medidas por los detectores de partГӯculas en el acelerador.
- Las 7 caracterГӯsticas restantes son calculadas a partir de las 21 anteriores, y fueron derivadas por los fГӯsicos para facilitar el proceso de clasificaciГіn.

Cada dato estГЎ etiquetado en una de dos posibles caracterГӯsticas:

- "1": presencia de seГұal (posiblemente una partГӯcula)
- "0": ruido de fondo (no corresponde a una partГӯcula)

Para mГЎs informaciГіn se sugiere mirar los detalles en el artГӯculo [Searching for Exotic Particles in High-Energy Physics with Deep Learning](https://arxiv.org/pdf/1402.4735.pdf).

Teniendo en cuenta el elevado nГәmero de datos que contiene este set, para esta prГЎctica se trabajarГЎ con una versiГіn reducida del mismo (*HIGGS_subset.csv*) que contiene un total de 100.000 datos.

## 2.1- PreparaciГіn de Google Drive y lectura del set de datos

```python
import pandas as pd

ruta = 'https://raw.githubusercontent.com/cagomezv/AP_USA/refs/heads/main/Semana_1/Data/HIGGS_subset.csv'
df = pd.read_csv(ruta)
print(df.shape)
df.head()
```

```python
# DistribuciГіn de categorГӯas
df.iloc[:,0].value_counts()
```

##В 2.2- CreaciГіn del set de entrenamiento

```python
X = df.values[:,1:]
Y = df.values[:,0]
print(X.shape)
print(Y.shape)
```

# 3- CreaciГіn y entrenamiento del modelo

- Capa de entrada: 28 elementos
- Capa de salida: 1 elemento (neurona de salida con activaciГіn sigmoidal)

Cuando no se tiene un referente previo se sugiere crear un primer modelo lo mГЎs simple posible. Crearemos un modelo con **una** capa oculta y el nГәmero de neuronas el promedio de las capas de entrada y de salida:

$n_{neuronas} = \frac{28+1}{2} = 15$

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input
import tensorflow as tf
import numpy as np
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)

DIM_ENTRADA = X.shape[1]   # 28
DIM_SALIDA = 1

modelo = Sequential()
modelo.add(Dense(15, activation='relu', input_shape=(DIM_ENTRADA,)))
modelo.add(Dense(DIM_SALIDA, activation='sigmoid'))

modelo.summary()
```

Verifiquemos el nГәmero de parГЎmetros:

$ \text{Entrada a capa oculta: } 28x15 + 15 = 435$
$ \text{Capa oculta a salida: } 15x1 + 1 = 16$

```python
# CompilaciГіn y entrenamiento
from tensorflow.keras.optimizers import Adam

EPOCHS = 1000
TAM_LOTE = 10000

opt = Adam()
modelo.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])

historia = modelo.fit(X, Y, epochs=EPOCHS, batch_size = TAM_LOTE, verbose=1)
```

# 4 - Resultados

```python
import matplotlib.pyplot as plt

def graficar(historia):
  plt.figure(figsize=(15,5))

  loss = historia.history['loss']
  accuracy = historia.history['accuracy']
  epochs = historia.epoch

  plt.subplot(121)
  plt.plot(epochs,loss,label='train_loss')
  plt.xlabel('Epochs')
  plt.ylabel('Loss')
  plt.legend()

  plt.subplot(122)
  plt.plot(epochs,accuracy,label='train_acc')
  plt.xlabel('Epochs')
  plt.ylabel('Accuracy')
  plt.legend()
```

```python
graficar(historia)
```

#5 - Sugerencias adicionales


Se pueden afinar diferentes hiperparГЎmetros para intentar mejorar la precisiГіn del modelo:

1. Cambiando el *nГәmero de neuronas* de la capa oculta
2. Agregando mГЎs capas y cambiando el nГәmero de neuronas de cada capa
3. Incrementando el *nГәmero de epochs*
4. Cambiando el *batch_size*
